```python
import pandas as pd
import numpy as np

#1

file = 'mower_market_snapshot.csv'
df = pd.read_csv(file, sep=';')
df
```

```python
#3   Auditer et analyser la colonnes Prod_cost 

def diedValue(df, col):
    values = pd.to_numeric(df[col], errors="coerce")
    return df[values<0].append(df[values.isna()])

diedValue(df,"prod_cost")
```

```python
#4  Analyser le nombre de valeurs manquantes par colonnes 

def nullValues(df):
    return pd.isnull(df)

nullValues(df).sum()
```

```python
#5 Remplacer les valeurs manquantes par la moyenne de la colonne (si possible)


#df.fillna(df.mean())
def replaceOfMeans(df, col):
    df[col] = pd.to_numeric(df[col], errors="coerce")
    df.fillna(df[col].mean(), inplace=True)
    
replaceOfMeans(df, 'prod_cost')
df

```

```python
#6  Afficher les valeurs uniques de la colonne Warranty 

def showEveryUniqueValue(df, col):
    return df[col].unique()

print(showEveryUniqueValue(df, "warranty"))
```

```python
#7 8  Traiter la colonne warranty afin d'uniformiser les valeurs (avoir uniquement des valeurs numériques uniformes) 
#df["warranty"] = df["warranty"].apply(lambda x: x[0])


def modifyValue(df, col):
     df[col] = df[col].apply(lambda x: x[0])


modifyValue(df, "warranty")
df
```
